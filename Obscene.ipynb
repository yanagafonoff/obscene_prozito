```python
import pymystem3
import pandas as pd
```

```python
mystem = pymystem3.Mystem()
```

```python
texts_df = pd.read_csv('test2.csv', sep='\t')
texts_df
```

```python
meta_df = pd.read_csv('diary.csv', sep = ',#')
meta_df
```

```python
merged_df = texts_df.merge(meta_df, left_on='"diary"', right_on='"id"', suffixes=('left', ''))
clean_df = merged_df[['"id"left', '"text"', '"date"', '"notes"', '"person"']]
clean_df
```

```python
x = map(mystem.analyze, map(str,clean_df['"text"']))
result_dict = {'id': [], 'word': [], 'date': [], 'person': []}
for i in range(100000):
    analysis = next(x)
    for dct in analysis:
        key = 'analysis'
        if key in dct and len(dct[key]) > 0:
            word = dct['text']
            gr = dct[key][0]['gr']
            if 'обсц' in gr:
                date = clean_df.iloc[i]['"date"']
                idd = clean_df.iloc[i]['"id"left']
                person = clean_df.iloc[i]['"person"']
                print(i, word, gr, idd)
                result_dict['id'].append(idd)
                result_dict['word'].append(word)
                result_dict['date'].append(date)
                result_dict['person'].append(person)
                
pd.DataFrame(data=result_dict)
```
